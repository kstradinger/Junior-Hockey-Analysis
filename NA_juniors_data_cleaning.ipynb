```python
#Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
#Import CSV
juniors_df=pd.read_csv('data_sets/na_juniors_99to19.csv')
juniors_df.head()
```

```python
# Drop columns that will not be used & duplicates
juniors_df=juniors_df.drop(['DOB'],axis=1)

juniors_df=juniors_df.drop_duplicates(keep='last', subset='PLAYER_ID')
```

```python
#player count of all players which we have juniors hockey data for 
players = juniors_df['PLAYER_ID'].count()
print(f'Total unique players: {players}')
```

```python
# create a unique DF for only players who are drafted
drafted_df = juniors_df[juniors_df['DRAFTED']=='Yes']
drafted_df.head()
```

```python
# we have data for 2810 players who were drafted out of a total of pool of 13,997 players
drafted_players = drafted_df['PLAYER_ID'].nunique()
print(f'Total unique drafted players: {drafted_players}')
```

```python
#create unique df for each league
ohl_df=juniors_df[juniors_df['LEAGUE']=='OHL']

whl_df=juniors_df[juniors_df['LEAGUE']=='WHL']

ushl_df=juniors_df[juniors_df['LEAGUE']=='USHL']

qmjhl_df=juniors_df[juniors_df['LEAGUE']=='QMJHL']
```
