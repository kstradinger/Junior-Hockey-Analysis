```python
#Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
#Import CSV
juniors_df=pd.read_csv('na_juniors_99to19.csv')
juniors_df.head()
```

```python
# Drop columns that will not be used & duplicates
juniors_df=juniors_df.drop(['DOB'],axis=1)

juniors_df=juniors_df.drop_duplicates(keep='last', subset='PLAYER_ID')
```

```python
#player count and unique, these are the same because duplicates were dropped except for the last season in juniors
juniors_df['PLAYER_ID'].nunique()
# 13,997 Unique players
juniors_df['PLAYER_ID'].count()
#  35,251 entries
```

```python
# create a unique DF for only players who are drafted

drafted_df = juniors_df[juniors_df['DRAFTED']=='Yes']
drafted_df.head()
```

```python
# we have data for 2810 players who were drafted out of a total of pool of 13,997 players
drafted_df['PLAYER_ID'].nunique()
```
