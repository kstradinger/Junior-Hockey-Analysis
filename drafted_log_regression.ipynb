```python
# Function to convert League from string into int
def league_int(x):
    if x.LEAGUE == 'WHL':
        return 0
    elif x.LEAGUE == 'QMJHL':
        return 1
    elif x.LEAGUE == 'OHL':
        return 2
    elif x.LEAGUE == 'USHL':
        return 3
```
